## 2. Research design

I will be asking the following three questions:
#### 1. Does Relationship Satisfaction have any bearing on Job Satisfaction?
#### 2. Does Hourly rate have any bearing on Attrition?
#### 3. Is a higher number of years since an employee's last promotion a factor in attrition for employees who have long tenure with the company? 

The alternative hypothesis for all questions will be:
#### 1. Relationship satisfaction is a factor in job satisfaction
#### 2. Hourly rate is a factor in attrition
#### 3. A higher number of years since an employee's last promotion is a factor in attrition for employees who have long tenure with the company

**For Q1**, I will split the data into 4 categories based on the Relationship Satisfactin ratings 1-4 (Low, Medium, High, Very High) and use a Kruskal-Wallis test to see if there is a significant difference in Job Satisfaction ratings.

**For Q2**, I will split the data into 2 categories of employees with the company and those who have left. Then I will use a T-Test to see if there is a significant difference in Hourly Rates. Potentially a low paying job (little money) or a high paying job (too much stress) have higher attrition.

**For Q3**, I will filter the data at the 75%ile to study the longest-tenured employees. Then split the data by those who have left the company and those who are still with the company. Then I will use a Kruskal-Wallis test to see if there is a significant difference in the span of time since their last promotion to see if this was a factor in attrition.
